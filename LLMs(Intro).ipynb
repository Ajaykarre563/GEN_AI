# **What is LLMs?**

Large Language Models (LLMs) are deep learning models trained on extensive text data to understand and generate human-like text. They use advanced architectures like transformers to process language and can perform tasks like answering questions, summarizing text, translating, and more, mimicking human communication.

**LLMs are good understanding and generating human language**

# **LLMs model core architecture**

**The core architecture of most Large Language Models (LLMs) is the Transformer a type of netural network architecture invented by Google.**

llm model.png

**Variants of Transformer Architecture in LLMs**

**Decoder-Only (GPT):**
      
Uses only the decoder part of the Transformer.

Focuses on generating text and autoregressive tasks.

**Encoder-Only (BERT):**
        
Uses only the encoder part of the Transformer.

Designed for understanding tasks like classification and sentiment analysis.

**Encoder-Decoder (T5):**
        
Combines encoder and decoder.
      
Used for tasks like translation, summarization, and Q&A.

# **Encoder-Only Models:**

Encoder-only models focus on understanding input text by capturing contextual representations through bidirectional attention. These are primarily used for tasks like text classification, named entity recognition, and sentence embedding.

Encoder-only Models (Blue):

BERT - Bidirectional transformer for understanding the context in text.
    
DistilBERT - A lighter, faster, and smaller version of BERT.
    
RoBERTa - Enhanced BERT with robust optimization and more training data.
    
XLM - Cross-lingual pretraining model for multilingual tasks.
    
XLM-R - Improved multilingual model based on RoBERTa.
    
ALBERT - A lighter and faster variant of BERT with reduced parameters.
    
ELECTRA - Pretrained using a replaced token detection task for efficiency.
    
DeBERTa - Added disentangled attention and enhanced input embeddings for better understanding.

# **Decoder-Only Models:**

Decoder-only models are autoregressive, predicting one token at a time to generate coherent sequences. They are ideal for text generation tasks like story writing, chatbots, and code completion.

Decoder-only Models (Red):

GPT - Pioneered autoregressive language modeling with transformers.
    
GPT-2 - Scaled-up version of GPT, showcasing more coherent text generation.
    
CTRL - Model conditioned on control codes for guided text generation.
    
GPT-3 - Large-scale model with 175B parameters, excelling in diverse NLP tasks.
    
GPT-Neo - Open-source alternative to GPT-3, developed by EleutherAI.
    
GPT-J - EleutherAI’s model emphasizing open-source development.
    
GPT-4 - Enhanced multimodal understanding and reasoning capabilities.
    
ChatGPT - Specialized GPT-based conversational model optimized for dialogue.

## **Encoder-Decoder Models:**

These models combine encoders (to understand inputs) and decoders (to generate outputs), making them versatile for translation, summarization, and question-answering tasks. The encoder processes the input, and the decoder generates output based on the encoder’s representation.


Encoder-Decoder Models (Gray):

T5 - Flexible transformer for text-to-text tasks, like summarization and translation.
    
BART - Combines bidirectional encoding and autoregressive decoding for generation tasks.
    
M2M-100 - Facebook’s multilingual model for direct translation across 100 languages.
    
BigBird - Scalable transformer designed to handle longer sequences efficiently.
    
Llama-2 - A family of large language models optimized for performance and efficiency.
    
Gemini - Advanced multimodal model combining LLMs with visual understanding.
    
Bard - Google’s conversational AI leveraging LaMDA for real-time dialogue.

**Prompt engineering and prompt design are concepts related to working with AI systems, particularly those that use natural language processing models like GPT (Generative Pre-trained Transformer). They focus on crafting effective input queries (or prompts) to achieve the desired output from the AI model.**

# **Prompt Engineering:**

Prompt engineering is the process of creating and optimizing prompts to guide an AI model's behavior effectively. It involves:

Structuring prompts in a way that makes them clear and contextually relevant.
    
Experimenting with phrasing, examples, and instructions to improve the output quality.
    
Testing variations to find the best-performing input.

Example:

If you want the AI to summarize a text, the prompt might be:

**Basic Prompt:**
    
"Summarize the following text:"

**Engineered Prompt:**
    
"Summarize the following text in three bullet points, highlighting the main ideas and omitting unnecessary details:"

# **Prompt Design**

Prompt Design is about creating easy-to-use and clear instructions for users to interact with an AI. It ensures that the prompts are simple and help people get what they need without confusion.

**Example: Language Translation**

Poorly Designed Prompt:
    
"Translate this."

Well-Designed Prompt:
    
"Enter the text you want to translate and the language you need (e.g., English to Spanish)."

# **Large-Shot Learning**

Definition: The AI is trained with a large number of examples to learn a task.
    
**Example:**
    
Training a model to recognize cats by showing it thousands of cat images labeled with "cat."

# **Few-Shot Learning**

Definition: The AI learns a task with only a few examples.

**Example:**

Teaching an AI to recognize a new medical condition (e.g., a rare skin rash) by providing it with 3–5 labeled images of patients with the condition.
