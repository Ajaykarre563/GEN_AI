```python
from transformers import pipeline
```

**CUDA:** A platform by NVIDIA that enables GPU acceleration for computational tasks, such as deep learning.(i want to look my model Inside the GPU).

```python
text = "Python is a high-level, general-purpose programming language."

pipe = pipeline("text-to-speech", model="suno/bark-small", device="cuda")

output = pipe(text)
```

```python
output
```

**IPython.display import Audio means you are importing the Audio class from the IPython.display module.**

**The Audio class is used to embed and play audio directly in Jupyter notebooks. It supports various formats, including raw NumPy arrays, file paths, and byte streams.**

This makes it a convenient tool for interactive programming, particularly when dealing with audio data in machine learning, signal processing, or multimedia projects.

```python
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
