```python
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain.llms import OpenAI
```

```python
import pandas as pd
```

#**03: Setup the Environment**


```python
os.environ['OPENAI_API_KEY'] = ""
```

#**04:Data**

```python
url = "https://raw.githubusercontent.com/adamerose/datasets/master/titanic.csv"
df=pd.read_csv(url)
print(df.shape)
df.head()
```

#**05: Single DataFrame Example**

#**Agent can interact with Single DataFrame**

```python
llm=OpenAI()
```

```python
agent=create_pandas_dataframe_agent(llm, df, verbose=True, allow_dangerous_code=True)
```

```python
agent.run("How many rows are there")
```

```python
agent.run("How many people have more than 23 age")
```

**A multi-dataframe agent is a tool or system designed to work with multiple dataframes (structured datasets, often tables) at the same time. Instead of focusing on just one dataframe, it can analyze, compare, or process information across several dataframes in parallel or sequentiall**`

**For example:**

If you have two dataframes, one with customer data and another with sales data, a multi-dataframe agent could:
        
Match customers to their sales records.

Analyze trends by combining information from both tables.

Think of it like a helper that understands how to connect and make sense of multiple pieces of tabular data.

#**06: Multi DataFrame Example**

#**Agent can also interact with Multiple DataFrames passed in a list**

```python
df.info()
```

```python
df1=df.copy()
```

```python
df1
```

```python
df1["age"] = df1["age"].fillna(df1["age"].mean())
```

```python
agent=create_pandas_dataframe_agent(llm, [df, df1], verbose=True,allow_dangerous_code=True)
```

```python
agent.run("How many rows in the age column are different")
```

```python
df2=df1.copy()
```

```python
df2["Age_Multiplied"]=df1["age"] * 2
```

```python
df2.head()
```

```python
agent=create_pandas_dataframe_agent(llm, [df,df1, df2], verbose=True, allow_dangerous_code=True)
```

```python
agent.run("Are the number of columns same in all the dataframe")
```
